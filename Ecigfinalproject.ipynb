### Part 1: Time-Based Plots for Flavored and Unflavored E-cigarette participants


We need to index the dataframe into two, one with participants who recieved flavored e-cigarretes, `flavored_data` (marked by `0`), and those who recieved unflavored cigarettes, `unflavored_data` (marked by `1`). These numbers are marked in the `condition` column of the original dataset.

As well, participants marked by each row in the dataset, are the same when print the shape of each group, so the results we find will be based on the same amount of participants.

```python
unflavored_data = ecig_data[ecig_data['condition'] == 0]
flavored_data = ecig_data[ecig_data['condition'] == 1]

print(unflavored_data.shape)
print(flavored_data.shape)
```

Based on participants' sweet craving, and savoury craving, and total cigarettes each day, the columns pertaining to each element are marked as strings within a list, being `sweet_column`, `savoury _column`, and `cigarette_column`.

```python
sweet_column = ['daily_1_sweetcraving','daily_2_sweetcraving','daily_3_sweetcraving',
                      'daily_4_sweetcraving','daily_5_sweetcraving','daily_6_sweetcraving',
                      'daily_7_sweetcraving']

savoury_column =   ['daily_1_savourycraving','daily_2_savourycraving','daily_3_savourycraving',
                      'daily_4_savourycraving','daily_5_savourycraving','daily_6_savourycraving',
                      'daily_7_savourycraving']

cigarette_column = ['daily_1_cigarettetotal','daily_2_cigarettetotal','daily_3_cigarettetotal',
                        'daily_4_cigarettetotal','daily_5_cigarettetotal', 'daily_6_cigarettetotal',
                      'daily_7_cigarettetotal']
```

Now, we input each component's list into our `flavored_data`. Then, we take the mean value recorded of that component, documented as an `average_flavored_` variable. Sweet and savoury cravings are marked on a scale of 1-7, and total cigarettes are any documented number. Then, we can see the mean value of each component for each day of the experiment.

```python
flavored_sweet_data = flavored_data[sweet_column]
average_flavored_sweet = (flavored_sweet_data.mean())
print(average_flavored_sweet)

flavored_cigarette_data = flavored_data[cigarette_column]
average_flavored_cigarette = (flavored_cigarette_data.mean())
print(average_flavored_cigarette)

flavored_savoury_data = flavored_data[savoury_column]
average_flavored_savoury = (flavored_savoury_data.mean())
print(average_flavored_savoury)
```

We repeat the code, but now inserting each component's list into the unflavored data. 

```python
unflavored_sweet_data = unflavored_data[sweet_column]
average_unflavored_sweet = (unflavored_sweet_data.mean())
print(average_unflavored_sweet)

unflavored_savoury_data = unflavored_data[cigarette_column]
average_unflavored_savoury = (unflavored_savoury_data.mean())
print(average_flavored_savoury)

unflavored_cigarette_data = unflavored_data[savoury_column]
average_unflavored_cigarette = (unflavored_cigarette_data.mean())
print(average_unflavored_cigarette)
```

### Part 2: End-Of-Experiment Plot for Flavored and Unflavored E-cigarette Participants

We will look at total vape duration (in seconds), as well as find the total amount of agreeable statements when asked about addiction for each group.

Similarly to the previous code, we will find the mean value of the data recorded, but now in only one column, being the `vape_duration_s`. We plug in this column into each dataset separately, and find the mean value of time for each group. This is recorded as the total time spent vaping throughout the experiment.

```python
unflavored_vape_time = unflavored_data['vape_duration_s']
print(unflavored_vape_time.mean())

flavored_vape_time = flavored_data['vape_duration_s']
print(flavored_vape_time.mean())
```

10 questions are asked to participants at the end of the experiment. 

1. "I have a desire for a cigarette right now"
2. "Nothing would be better than smoking a cigarette right now"
3. "If it were possible, I probably would smoke right now"
4. "I could control things better right now if I could smoke"
5. "All I want right now is a cigarette"
6. "I have an urge for a cigarette"
7. "A cigarette would taste good now"
8. "I would do almost anything for a cigarette now"
9. "Smoking would make me less depressed"
10. "I am going to smoke as soon as possible"

The participant responses were based on these metrics:
**1** = strongly disagree;
**2** = disagree;
**3** = somewhat disagree;
**4** = neither agree nor disagree;
**5** = somewhat agree; **6** = agree; **7** = strongly agree. 5, 6, and 7 will be marked as the agreeable statements.

To measure the amount of agreeable statements, we create a list, `column_questions`, and insert each question-labeled column, as strings. Then we create a 'for loop', that looks for the the numbers: 5, 6, or 7 in each dataset. If they do appear, then each number is counted as one, and summed up for each group.

```python
column_questions = ['QSU_1_S2post','QSU_2_S2post','QSU_3_S2post','QSU_3_S2post',  
                        'QSU_5_S2post','QSU_6_S2post','QSU_7_S2post','QSU_8_S2post',
                        'QSU_9_S2post','QSU_10_S2post']

unflavored_counter = 0

for column in column_questions:
    unflavored_counter += unflavored_data[column].isin([5, 6, 7]).sum()

print(unflavored_counter)


flavored_counter = 0 
for column in column_questions:
    flavored_counter += flavored_data[column].isin([5, 6, 7]).sum()

print(flavored_counter)                                                        
```

### Part 3: Time-Based and End-of-Experiment Plots for Flavored E-cigarette Combinations
Since flavored e-cigarette participants are given two different flavors, we will investigate which combination creates a greater response, using the metrics from parts 1 and 2. Here, we might be able to find a correlation for one individual flavor being more potent towards addiction.

Each flavor combination is given its own list, and each flavor is marked with a number in the dataframe. We created a 'for loop', where if columns of `flavour 1` and `flavour 2` have numbers that correspond to a specific combo, that row is added to that combo's list. Either number is capable of appearing in either column, hence the use of `or` and `and` as conditions in our loop. We then convert each list into a dataframe to perform the proceeding functions.

```python
blackcurrant_strawberry = []
blackcurrant_vanilla = []
blackcurrant_caramel = []
strawberry_vanilla = []
strawberry_caramel = []
caramel_vanilla = []


blackcurrant = 1  
strawberry = 2  
vanilla = 3
caramel = 4


for index, row in flavored_data.iterrows():
    if ((row['flavour1'] == blackcurrant or row['flavour2'] == blackcurrant) and
        (row['flavour1'] == strawberry or row['flavour2'] == strawberry)):
        blackcurrant_strawberry.append(row)
blackcurrant_strawberry_df = pd.DataFrame(blackcurrant_strawberry)                     

for index, row in flavored_data.iterrows():
    if ((row['flavour1'] == blackcurrant or row['flavour2'] == blackcurrant) and
        (row['flavour1'] == vanilla or row['flavour2'] == vanilla)):
        blackcurrant_vanilla.append(row)
blackcurrant_vanilla_df = pd.DataFrame(blackcurrant_vanilla)

for index, row in flavored_data.iterrows():
    if ((row['flavour1'] == blackcurrant or row['flavour2'] == blackcurrant) and
        (row['flavour1'] == caramel or row['flavour2'] == caramel)):
        blackcurrant_caramel.append(row)
blackcurrant_caramel_df = pd.DataFrame(blackcurrant_caramel)

for index, row in flavored_data.iterrows():
    if ((row['flavour1'] == strawberry or row['flavour2'] == strawberry) and
        (row['flavour1'] == vanilla or row['flavour2'] == vanilla)):
        strawberry_vanilla.append(row)
strawberry_vanilla_df = pd.DataFrame(strawberry_vanilla)

for index, row in flavored_data.iterrows():
    if ((row['flavour1'] == strawberry or row['flavour2'] == strawberry) and
        (row['flavour1'] == caramel or row['flavour2'] == caramel)):
        strawberry_caramel.append(row)
strawberry_caramel_df = pd.DataFrame(strawberry_caramel)

for index, row in flavored_data.iterrows():
    if ((row['flavour1'] == caramel or row['flavour2'] == caramel) and
        (row['flavour1'] == vanilla or row['flavour2'] == vanilla)):
        caramel_vanilla.append(row)
caramel_vanilla_df = pd.DataFrame(caramel_vanilla)
```
